```python
neows_api = "https://api.nasa.gov/neo/rest/v1/feed?"
response = requests.get(neows_api, params={'api_key': api_key})
print(response.json())
```

```python
data = response.json()

data = data['near_earth_objects']
```

```python
final_data = {
    'Asteroid ID' : [],
    'Asteroid name' : [],
    'The Minimal estimated diameter in Kilometre' : [], 
    'Absolute_magnitude' : [],
    'Relative_velocity(km/s)' : []
}

for key in data.keys():
    for obj in data[key]:
        final_data['Asteroid ID'].append(obj['id'])
        final_data['Asteroid name'].append(obj['name'])
        final_data['The Minimal estimated diameter in Kilometre'].append(obj['estimated_diameter']['kilometers']['estimated_diameter_min'])
        final_data['Absolute_magnitude'].append(obj['absolute_magnitude_h'])
        final_data['Relative_velocity(km/s)'].append(obj['close_approach_data'][0]['relative_velocity']['kilometers_per_second'])
```

```python
df = pd.DataFrame(final_data,)
df
```

```python
df.to_csv('Asteroid_dataset', index=False)
```
